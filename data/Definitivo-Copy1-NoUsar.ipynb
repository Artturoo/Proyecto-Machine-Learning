Modificamos la columna job_title. Nos quedamos con 4 variables

```python
def job_title(x):
    if 'Machine' in x or 'ML' in x or 'Computer' in x:
        return 'Machine Learning'
    if 'Engineer' in x or 'Engineering' in x:
        return 'Data Engineer'
    if 'Analyst' in x or 'Analytics' in x:
        return 'Data Analyst'
    else:
        return 'Data Scientist'
```

```python
salary.job_title=salary.job_title.apply(job_title)
```

```python
testeo.job_title=testeo.job_title.apply(job_title)
```

Reemplazamos los valosres CT y FL por PT y FT respectivamente para quedarnos con esas 2 variables unicamente

```python
salary.employment_type = salary.employment_type.replace('CT', 'PT')
salary.employment_type = salary.employment_type.replace('FL', 'FT')
```

```python
testeo.employment_type = testeo.employment_type.replace('CT', 'PT')
testeo.employment_type = testeo.employment_type.replace('FL', 'FT')
```

En la columna company_location nos quedamos solamente con los 10 primeros que son ma mayoria de paises

```python
def company_location(x):
    if 'US' in x:
        return 'US'
    elif 'GB' in x:
        return 'GB'
    elif 'CA' in x:
        return 'CA'
    elif 'DE' in x:
        return 'DE'
    elif 'IN' in x:
        return 'IN'
    elif 'FR' in x:
        return 'FR'
    elif 'ES' in x:
        return 'ES'
    elif 'GR' in x:
        return 'GR'
    elif 'JP' in x:
        return 'JP'
    else:
        return 'Other'
```

```python
salary.company_location=salary.company_location.apply(company_location)
```

```python
testeo.company_location=testeo.company_location.apply(company_location)
```

```python
salary_x=salary.drop('salary_in_usd', axis=1)
salary_y=salary.salary_in_usd
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(salary_x, salary_y,train_size=0.9, test_size=0.1,random_state =71)
```

```python
from sklearn.linear_model import LinearRegression as LinReg
from sklearn.linear_model import Lasso        # regularizacion L1
from sklearn.linear_model import Ridge        # regularizacion L2
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor as RFR  
from sklearn.tree import ExtraTreeRegressor as ETR
from sklearn.ensemble import GradientBoostingRegressor as GBR
from xgboost import XGBRegressor as XGBR
from catboost import CatBoostRegressor as CTR
```

```python
from lightgbm import LGBMRegressor as LGBMR
```

**Iniciamos los modelos**

```python
linreg=LinReg()
lasso=Lasso()
ridge=Ridge()
elastic=ElasticNet()
svr=SVR()
rfr=RFR()
etr=ETR()
gbr=GBR()
xgbr=XGBR()
ctr=CTR()
lgbmr=LGBMR()
```

**Entrenamos modelos**

```python
linreg.fit(X_train, y_train)
lasso.fit(X_train, y_train)
ridge.fit(X_train, y_train)
elastic.fit(X_train, y_train)
svr.fit(X_train, y_train)
rfr.fit(X_train, y_train)
etr.fit(X_train, y_train)
gbr.fit(X_train, y_train)
xgbr.fit(X_train, y_train)
ctr.fit(X_train, y_train)
lgbmr.fit(X_train, y_train)
```
